```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import butter,filtfilt
df = pd.read_csv('https://raw.githubusercontent.com/AfrizalSY/IoT_LPF-MVA/main/12.csv',
                 names=['#','x','y','z','*'])
def butter_lowpass_filter(data, cutoff, order):
    b, a = butter(order, cutoff, btype='lowpass', analog=False,
                  output="ba")
    y = filtfilt(b, a, data)
    return y
df['x'].plot(figsize=(16,6), label="real",title="LPF")
x = butter_lowpass_filter(df['x'],cutoff=0.1,order=2)
df = pd.DataFrame(x)
df[0].plot(label="data x")
plt.legend()
```
